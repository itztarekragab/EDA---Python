## Imports

```python
import pandas as pd
```

## Reading data

```python
df1 = pd.read_csv('user_usage.csv')
df2 = pd.read_csv('user_device.csv')
```

## Showing the two data frames

```python
df1
```

```python
df2
```

## Using merge

> Outer merge

```python
pd.merge(how='outer',left=df1,right=df2,on='use_id')
```

> Inner merge

```python
pd.merge(how='inner',left=df1,right=df2,on='use_id')
```

> Left merge

```python
pd.merge(how='left',left=df1,right=df2,on='use_id')
```

> Right merge

```python
pd.merge(how='right',left=df1,right=df2,on='use_id')
```

## Using join

```python
# we need to work on index so we edit the dataframes

df1 = pd.read_csv('user_usage.csv')
df2 = pd.read_csv('user_device.csv',index_col='use_id')
```

> Outer join

```python
df1.join(df2,how='outer',on='use_id')
```

> Inner join

```python
df1.join(df2,how='inner',on='use_id')
```

> Left join

```python
df1.join(df2,how='left',on='use_id')
```

> Right join

```python
df1.join(df2,how='right',on='use_id')
```
