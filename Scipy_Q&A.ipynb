
## Draw a line in a diagram from position (1,2) to position (10,20)
 - 1 set the plot title >> "points plot" , xlabel: "x-axis" , ylabel : "y-axis"
 - 2 using marker type  "star" to represent the points and marker color :red|

```python
import matplotlib.pyplot as plt
import numpy as np

points = np.array([[1,2],
                  [10,20]])
plt.plot(points[:,0],points[:,1],"*r-",)
plt.savefig("firstplot.png")
plt.show()
```

## Get the local min of the function below using scipy. then plot the equation with the local min point using matlibplot


```python
from scipy.optimize import minimize

def f(x):
    return x**2 + x + 2

minimum = minimize(f,0,method='BFGS')
print(minimum.x)

x_values = np.arange(-10,10,0.1)
y_values = np.apply_along_axis(f,arr=x_values,axis=0)
plt.plot(x_values,y_values,"g")
plt.plot(minimum.x,f(minimum.x),'*r')
plt.show()
```

## Compute the solution of the system of equations using scipy




```python
from scipy.linalg import solve
a=np.array([[2,1],
            [1,1]])
b=np.array([1,1])

print(solve(a,b))
```

```python
from scipy.interpolate import UnivariateSpline,Rbf

x = np.arange(0,360,5)
y = np.sin(np.deg2rad(x))**2-np.cos(np.deg2rad(x))**2
spl = UnivariateSpline(x,y)
spl.set_smoothing_factor(0.01)
rbf = Rbf(x,y)
x_new = np.arange(0,360,10)
plt.plot(x_new,spl(x_new),'g',x,y,'ro',ms=5)
plt.show()
plt.plot(x_new,rbf(x_new),'g',x,y,'ro',ms=5)
plt.show()
# The Two of them are almost identical when smoothing on 0.01
```
